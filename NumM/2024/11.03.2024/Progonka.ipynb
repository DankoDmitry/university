```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x): return np.sin(x)

def analit(x): return -1*np.sin(x)

X_0 = np.pi/2
X_n = 3*np.pi/2
n = 1000
h = (X_n - X_0) / n

def F(i): return np.sin(X_0 + h*(i-1))

Xi = np.zeros(n)
Eta = np.zeros(n+1)

Xi[1] = 1/(2)
for i in range(2, n-1):
    Xi[i+1] = 1 / (2 - Xi[i])

Eta[1] = h*h*F(1)/(2)
for i in range(2, n):
    Eta[i+1] = (Eta[i]-h*h*F(i)) / (2 - Xi[i])

y = np.zeros(n)
y[n-1] = 1

for i in range(n-2, -1, -1):
    y[i] = Xi[i+1]*y[i+1]+Eta[i+1]

t = np.linspace(X_0, X_n, n)

plt.plot(t, analit(t), color = 'blue', label = 'аналитическое решение')
plt.plot(t, y, color = 'green', label = 'численнле решение')
plt.legend()
plt.show()
```
