## Система *хищник-жертва*

```python
from scipy.integrate import odeint
from scipy.integrate import solve_ivp

import matplotlib.pyplot as plt

%matplotlib inline

import numpy as np
```

```python
u0 = [0.5, 1.1, 0.7]
v0 = [0.5, 2, 1]


gamma = .2
t0 = 0
tf = 100
st = .1

def lotkavolterra(t, z):
    u, v = z
    return [u*(1-v), gamma*v*(u - 1)]

sol = [solve_ivp(lotkavolterra, [t0, tf], [u0[i], v0[i]], max_step=st) for i in range(3)]

t = np.arange(t0-2*st, tf, st)


fig, ax = plt.subplots(3, 2, sharex=True)

for i in range(3):
    ax[i, 0].plot(t, (sol[i].y)[0], color="blue", label='жертва')
    ax[i, 0].plot(t, (sol[i].y)[1], color="red", label='жертва')

    ax[i, 1].plot((sol[i].y[0]), (sol[i].y)[1])

    ax[i, 0].legend()


plt.show()
```

```python
sol = solve_ivp(lotkavolterra, [t0, tf], [u0[2], v0[2]], max_step=st)

t = np.linspace(0, 100, 1002)



plt.plot(t, (sol.y)[0], color="blue", label='жертва')
plt.plot(t, (sol.y)[1], color="red", label='хищьник')


plt.xlabel('t')
plt.title('Lotka-Volterra System')
plt.show()
```
