```python
import math

Q = int(input())
R = int(input())

def find_divisors(n):
    divisors = []
    for i in range(1, int(math.sqrt(n))+1):
        if n % i == 0:
            divisors.append(i)
            if i != n // i:
                divisors.append(n // i)
    return len(divisors) 

print (find_divisors(Q*(R-1)))
```
