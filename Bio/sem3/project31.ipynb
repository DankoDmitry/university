```python
def pend(y, t, r, K):
    N = y
    dydt = r*N*(1-N/K)
    return dydt

r = 0.1
K = 1000

y0 = [10, 200, 700]

t = np.linspace(0, 20, 1000)

from scipy.integrate import odeint
sol = odeint(pend, y0, t, args=(r, K))

analit = (y0[0] * np.e ** (r*t)) / (1 - y0[0]/K + (y0[0] * np.e ** (r*t)) / K)
```

```python
def Xplicit_Y(y, tau, r, K, n0):
    if y == 0:
        return n0
    else:
        return (tau*r*y - tau*r*y*y/K + y)
```

```python
m = [10]

for i in range(1999):
    m.append(Xplicit_Y(m[i], 0.1, r, K, y0[0]))

# print(m)

plt.plot(m, 'g', label='N(0) = 10')
plt.plot(t, analit, 'b', label='N(0) = 10')

plt.show()
```

```python
# n - количество итераций, h - шаг, (x, y) - начальная точка
def Euler(n = 10, h = 0.01, x = 1, y = 1):
    for i in range(n):
        y += h * function(x, y)
        x += h
    return x, y # решение

def function(x, y):
    return 6 * x**2 + 5 * x * y # функция первой производной

print(Euler())
```
