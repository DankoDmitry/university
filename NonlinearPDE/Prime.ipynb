```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 99
N = 1
h = 1/((k+1))
t = np.arange(0, 1, (1/((k+1)*N)))


c = np.ones(k)
```

```python
def f(x):
    return np.sin(x)
```

```python
U = np.zeros(((k+1)*N))

U = f(t)
```

```python
print(U)
```

```python

plt.plot(t, U)
plt.show()
```

```python
v = np.zeros(((k+1)*N))

v[0] = U[1] - U[0]

for i in range(1, len(U)-1):
    v[i] = U[i+1] - U[i]

v[-1] = U[-1] - U[-2]

v = v/h

print(v)

plt.plot(t, v)
plt.show()
```

```python
u = np.zeros(((k+1)*N))

u[0] = v[1] - v[0]

for i in range(1, len(v)-1):
    u[i] = v[i+1] - v[i]

u[-1] = v[-1] - v[-2]

u = u/h

print(u)

plt.plot(t, u)
plt.show()
```
